## DataFrame

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Ahmed", "Sara", "  John", "Fatima", "Ali", "Ayesha", None,
             "Usman", "Zara", "Bilal", "Ahmed ", "Sana", "Hassan", "Ali",
             "Kiran", "Omar", "Fatima", "Zain", "Sara"],

    "Age": [22, 25, np.nan, 28, "23", 22, 21, 24,
            None, 26, 27, 25, "twenty", 30, 22,
            23, np.nan, 23, 29, 24],

    "City": ["Lahore", "Karachi", "lahore", "Islamabad", None, "Karachi", "Lahore", "Karachi",
             "Islamabad", "Lahore", "Karachi", "Karachi ", "Lahore", "Islamabad", "Lahore",
             "Karachi", "Lahore", "Islamabad", "Karachi", None],

    "Salary": [50000, 60000, 55000, None, 52000, 50000, "48000", 62000,
               61000, None, 58000, 60000, 57000, "65000", 50000,
               54000, None, 52000, 63000, 56000],

    "Department": ["IT", "HR", "Finance", "IT", "HR", "IT", None, "Finance",
                   "IT", "HR", "Finance", "HR", "IT", "Finance", "IT",
                   "HR", "Finance", None, "IT", "HR"]
}

df = pd.DataFrame(data)
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
```

```python
df["Name"] = df["Name"].str.strip()
df["City"] = df["City"].str.strip()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df['Age'].mean,inplace=True)
df["Salary"].fillna(df['Salary'].median,inplace=True)
```

```python
df["City"].fillna(df["City"].mode()[0], inplace=True)
df["Department"].fillna(df["Department"].mode()[0], inplace=True)
```

```python
df=df.drop(['Name'],axis=1)
```

```python
df.notnull().sum()
```
